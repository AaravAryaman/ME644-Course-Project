```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Data Preparation
temperatures = np.array([30, 30, 30, 30, 40, 40, 40, 50, 50, 50, 60, 60, 60, 60])
hardness = np.array([55.8, 59.1, 54.8, 54.6, 43.1, 42.2, 45.2, 31.6, 30.9, 30.8, 17.5, 20.5, 17.2, 16.9])
```

```python
# Compute linear hypothesis
slope, intercept, r_value, p_value, std_err = linregress(temperatures, hardness)
print(f"Linear Hypothesis: Hardness = {slope:.2f} * Temperature + {intercept:.2f}")
```

```python
# Plot the data and the linear hypothesis line
plt.figure
plt.scatter(temperatures, hardness, color='blue', label='Data')
plt.plot(temperatures, intercept + slope * temperatures, color='red', label='Linear Hypothesis')
plt.title('Heat Treatment Temperature vs Rockwell Hardness')
plt.xlabel('Temperature (°C)')
plt.ylabel('Hardness')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute coefficient of determination
r_squared = r_value**2
print(f"Coefficient of Determination: {r_squared:.4f}")
```
